```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

df_bina = df_bina.copy()

X = df_bina.drop("Çıxarış", axis=1)
y = df_bina["Çıxarış"]

y = y.map({"var": 1, "yoxdur": 0})

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])

estimator = LogisticRegression(max_iter=500, solver='liblinear')

full_pipeline_with_estimator = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('estimator', estimator)
])


param_grid = {
    'estimator__C': [0.01, 0.1, 1, 10],
    'estimator__solver': ['liblinear', 'lbfgs']
}

grid_search = GridSearchCV(full_pipeline_with_estimator, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)


print(f"Ən yaxşı parametrlər: {grid_search.best_params_}")


best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
print(f"Precision: {precision_score(y_test, y_pred):.4f}, Recall: {recall_score(y_test, y_pred):.4f}, F1-score: {f1_score(y_test, y_pred):.4f}, Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

df_bina = df_bina.copy()

X = df_bina.drop("Çıxarış", axis=1)
y = df_bina["Çıxarış"].map({"var": 1, "yoxdur": 0})

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median'))
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])


log_reg = LogisticRegression(max_iter=500, solver='liblinear')
xgb_clf = XGBClassifier(n_estimators=100, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss')

voting_clf = VotingClassifier(estimators=[('lr', log_reg), ('xgb', xgb_clf)], voting='hard')

full_pipeline_with_estimator = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('estimator', voting_clf)
])

param_grid = {
    'estimator__xgb__n_estimators': [500,600],
    'estimator__lr__C': [0.01,0.1, 0.2]
}

grid_search = GridSearchCV(full_pipeline_with_estimator, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Ən yaxşı parametrlər: {grid_search.best_params_}")

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall: {recall_score(y_test, y_pred):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred):.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def remove_outliers(df, num_features):
    Q1 = df[num_features].quantile(0.25)
    Q3 = df[num_features].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[~((df[num_features] < lower_bound) | (df[num_features] > upper_bound)).any(axis=1)]

df_bina = df_bina.copy()

num_features = df_bina.select_dtypes(include=[np.number]).columns
df_bina_cleaned = remove_outliers(df_bina, num_features)

X = df_bina_cleaned.drop("Çıxarış", axis=1)
y = df_bina_cleaned["Çıxarış"].map({"var": 1, "yoxdur": 0})

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median'))
])

cat_features = X_train.select_dtypes(exclude=[np.number]).columns

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])

log_reg = LogisticRegression(max_iter=500, solver='liblinear')
xgb_clf = XGBClassifier(n_estimators=100, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss')

voting_clf = VotingClassifier(estimators=[('lr', log_reg), ('xgb', xgb_clf)], voting='hard')

full_pipeline_with_estimator = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('estimator', voting_clf)
])

param_grid = {
    'estimator__xgb__n_estimators': [600,800],
    'estimator__lr__C': [0.01, 0.1, 0.2]
}

grid_search = GridSearchCV(full_pipeline_with_estimator, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Ən yaxşı parametrlər: {grid_search.best_params_}")

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall: {recall_score(y_test, y_pred):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred):.4f}")
```
